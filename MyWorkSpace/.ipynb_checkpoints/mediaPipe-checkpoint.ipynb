# Setup

## Import

```python
import cv2
import mediapipe as mp
import itertools
import numpy as np
from time import time
import matplotlib.pyplot as plt
```

## Setup MediaPipe

```python
# Inicia o mediapipe face mesh
mp_face_mesh = mp.solutions.face_mesh

# Ajusta os landmarks do rosto para imagens
face_mesh_images = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=2, min_detection_confidence=0.5)

# Ajusta os landmarks do rosto para videos
face_mesh_videos = mp_face_mesh.FaceMesh(static_image_mode=False, max_num_faces=1, min_detection_confidence=0.5,min_tracking_confidence=0.5)

# Inicia a classe de desenho do mediapipe
mp_drawing = mp.solutions.drawing_utils

# Inicia a classe de estilos de desenho do mediapipe
mp_drawing_styles = mp.solutions.drawing_styles
```

## Função auxilar

```python
def detectFacialLandmarks(image, face_mesh, display = True):
    # Converte para o formato RGB e coloca as landmarks
    results = face_mesh.process(image[:,:,::-1])
    
    # Cria uma cópia da imagem de entrada para desenhar as landmarks
    output_image = image[:,:,::-1].copy()
    
    # Verifica se encontrou as landmarks
    if results.multi_face_landmarks:

        # Repete para cada face encontrada
        for face_landmarks in results.multi_face_landmarks:

            # Aplica FACEMESH TESSALATION
            mp_drawing.draw_landmarks(image=output_image, landmark_list=face_landmarks,
                                      connections=mp_face_mesh.FACEMESH_TESSELATION,
                                      landmark_drawing_spec=None, 
                                      connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_tesselation_style())

            # Aplica FACEMESH CONTOURS
            mp_drawing.draw_landmarks(image=output_image, landmark_list=face_landmarks,
                                      connections=mp_face_mesh.FACEMESH_CONTOURS,
                                      landmark_drawing_spec=None, 
                                      connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_contours_style())
    if display:
        # Retorna imagem antes e depois
        plt.figure(figsize=[15,15])
        plt.subplot(121);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis('off');
        plt.subplot(122);plt.imshow(output_image);plt.title("Output");plt.axis('off');
    else:
        # Retorna a saida da imagem no formato BGR e os resultados da detecção do landmarks
        return np.ascontiguousarray(output_image[:,:,::-1], dtype=np.uint8), results              
```

# Image

```python
sample_img = cv2.imread('../Exemplo/media/sample.jpg')
```

### Demonstração

```python
detectFacialLandmarks(sample_img, face_mesh_images, display=True)
```

### Dados

```python
image_BGR, results_landmarks = detectFacialLandmarks(sample_img, face_mesh_images, display=False)
print(image_BGR)
```

# WebCam

```python
# Inicia o objeto da câmera 
camera_video = cv2.VideoCapture(0)
camera_video.set(3,1280)
camera_video.set(4,960)

# Cria uma janela para redimensionamento
cv2.namedWindow('Emotion Detection', cv2.WINDOW_NORMAL)

# Inicia o time
time1 = 0

# Fica tentando abrir a WebCam.
while camera_video.isOpened():
    
    # Lê os frames
    ok, frame = camera_video.read()

    # Checa se o frame não foi lido, então continua para proxima interação
    if not ok:
        continue
    
    # Inverte a matriz dos frames
    frame = cv2.flip(frame, 1)
    
    # Utiliza função para detectar os Landmarks do rosto
    frame, _ = detectFacialLandmarks(frame, face_mesh_videos, display=False)
    
    # Define o tempo
    time2 = time()
    
    #  Checagem para evitar divisão por 0
    if (time2 - time1) > 0:
    
        # Frames por segundo
        frames_per_second = 1.0 / (time2 - time1)
        
        # Escreve o número de frames
        cv2.putText(frame, 'FPS: {}'.format(int(frames_per_second)), (10, 30),
                    cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 3)
    
    # Atualiza os tempos do frame
    time1 = time2
    
    # Dispara o frame
    cv2.imshow('Emotion Detection', frame)
    
    # Pega a tecla pressionada 
    k = cv2.waitKey(1) & 0xFF    
    
    # Se a teclar for ESC, ele fecha 
    if(k == 27):
        break

# Fecha a window               
camera_video.release()
cv2.destroyAllWindows()
```
